```python
import numpy as np
import cv2

# Load the Haar Cascade for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Start capturing video
cap = cv2.VideoCapture(0)
cap.set(3, 640)  # Set width
cap.set(4, 480)  # Set height

try:
    while True:
        ret, img = cap.read()
        if not ret:
            print("Failed to grab frame")
            break

        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(
            gray,     
            scaleFactor=1.2,
            minNeighbors=5,     
            minSize=(20, 20)
        )

        for (x, y, w, h) in faces:
            # Draw a rectangle around the face
            cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)

        cv2.imshow('Video', img)

        # Check if the 'ESC' key is pressed or if the window is closed
        if cv2.waitKey(30) & 0xFF == 27:  # ESC key
            break
        if cv2.getWindowProperty('Video', cv2.WND_PROP_VISIBLE) < 1:
            break
except Exception as e:
    print(f"An error occurred: {e}")
finally:
    # Ensure proper cleanup
    cap.release()
    cv2.destroyAllWindows()
    print("Resources released, windows closed.")
```

```python
import cv2
import numpy as np

# Start capturing video
cap = cv2.VideoCapture(0)
cap.set(3, 640)  # Set width
cap.set(4, 480)  # Set height

# Variable to track the last captured frame
last_captured_image = None

try:
    while True:
        ret, img = cap.read()
        if not ret:
            print("Failed to grab frame")
            break

        # Convert the frame to grayscale
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Apply Gaussian blur to reduce noise
        blurred = cv2.GaussianBlur(gray, (5, 5), 0)

        # Threshold the image to detect contours
        _, thresh = cv2.threshold(blurred, 70, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

        # Find contours in the thresholded frame
        contours, _ = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

        # Count the number of significant contours
        finger_count = 0
        for contour in contours:
            if cv2.contourArea(contour) > 3000:  # Filter by area size to avoid noise
                finger_count += 1

        # Capture and display the image if exactly two fingers are detected
        if finger_count == 2:
            if last_captured_image is None or not np.array_equal(last_captured_image, img):
                last_captured_image = img.copy()
                cv2.imwrite("temporary_image.jpg", img)
                temp_img = cv2.imread("temporary_image.jpg")
                cv2.imshow("Captured Image", temp_img)

        # Display the video feed
        cv2.imshow('Video', img)

        # Exit conditions: ESC key or window close button
        if cv2.waitKey(30) & 0xFF == 27:  # ESC key
            break
        if cv2.getWindowProperty('Video', cv2.WND_PROP_VISIBLE) < 1:  # Window close button
            break

except Exception as e:
    print(f"An error occurred: {e}")
finally:
    # Ensure proper cleanup
    cap.release()
    cv2.destroyAllWindows()
    print("Resources released, windows closed.")
import cv2
import numpy as np

# Start capturing video
cap = cv2.VideoCapture(0)
cap.set(3, 640)  # Set width
cap.set(4, 480)  # Set height

# Variable to track the last captured frame
last_captured_image = None

try:
    while True:
        ret, img = cap.read()
        if not ret:
            print("Failed to grab frame")
            break

        # Convert the frame to grayscale
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Apply Gaussian blur to reduce noise
        blurred = cv2.GaussianBlur(gray, (5, 5), 0)

        # Threshold the image to detect contours
        _, thresh = cv2.threshold(blurred, 70, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

        # Find contours in the thresholded frame
        contours, _ = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

        # Count the number of significant contours
        finger_count = 0
        for contour in contours:
            if cv2.contourArea(contour) > 3000:  # Filter by area size to avoid noise
                finger_count += 1

        # Capture and display the image if exactly two fingers are detected
        if finger_count == 2:
            if last_captured_image is None or not np.array_equal(last_captured_image, img):
                last_captured_image = img.copy()
                cv2.imwrite("temporary_image.jpg", img)
                temp_img = cv2.imread("temporary_image.jpg")
                cv2.imshow("Captured Image", temp_img)

        # Display the video feed
        cv2.imshow('Video', img)

        # Exit conditions: ESC key or window close button
        if cv2.waitKey(30) & 0xFF == 27:  # ESC key
            break
        if cv2.getWindowProperty('Video', cv2.WND_PROP_VISIBLE) < 1:  # Window close button
            break

except Exception as e:
    print(f"An error occurred: {e}")
finally:
    # Ensure proper cleanup
    cap.release()
    cv2.destroyAllWindows()
    print("Resources released, windows closed.")
```
